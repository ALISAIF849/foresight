```python
import os
import requests
import time
import json
from typing import Optional, List, Dict
from pathlib import Path
from IPython.display import clear_output, display, Markdown

class ElevenLabsDubber:
    """
    Framework for dubbing videos using Eleven Labs API
    Optimized for Google Colab
    """

    def __init__(self, api_key: str, timeout: int = 300):
        """
        Initialize the Eleven Labs Dubber

        Args:
            api_key (str): Your Eleven Labs API key
            timeout (int): Request timeout in seconds (default: 300)
        """
        self.api_key = api_key
        self.base_url = "https://api.elevenlabs.io/v1"
        self.timeout = timeout

    def check_api_key(self) -> bool:
        """
        Verify that the API key is valid

        Returns:
            bool: True if API key is valid, False otherwise
        """
        try:
            headers = {"xi-api-key": self.api_key}
            response = requests.get(
                f"{self.base_url}/user",
                headers=headers,
                timeout=self.timeout
            )
            return response.status_code == 200
        except Exception as e:
            print(f"Error checking API key: {e}")
            return False

    def get_available_languages(self) -> Dict[str, str]:
        """
        Get list of available languages for dubbing
        Uses ISO 639-1 codes
        """
        languages = {
            'en': 'English', 'es': 'Spanish', 'fr': 'French', 'de': 'German',
            'it': 'Italian', 'pt': 'Portuguese', 'pl': 'Polish', 'nl': 'Dutch',
            'hu': 'Hungarian', 'ru': 'Russian', 'zh': 'Chinese (Mandarin)',
            'ja': 'Japanese', 'ko': 'Korean', 'tr': 'Turkish', 'sv': 'Swedish',
            'bg': 'Bulgarian', 'cs': 'Czech', 'da': 'Danish', 'el': 'Greek',
            'et': 'Estonian', 'fi': 'Finnish', 'hi': 'Hindi', 'lt': 'Lithuanian',
            'lv': 'Latvian', 'sk': 'Slovak', 'sl': 'Slovenian', 'ro': 'Romanian',
            'ar': 'Arabic', 'uk': 'Ukrainian', 'vi': 'Vietnamese', 'th': 'Thai',
            'id': 'Indonesian'
        }
        return languages

    def upload_video(self, video_path: str, target_language: str) -> Optional[str]:
        """
        Upload video file to Eleven Labs for processing with explicit MIME type.
        """
        try:
            if not os.path.exists(video_path):
                print(f"Error: Video file not found at {video_path}")
                return None

            file_size = os.path.getsize(video_path)
            print(f"üì§ Uploading video ({file_size / (1024*1024):.2f} MB)...")

            filename = os.path.basename(video_path)

            with open(video_path, 'rb') as f:
                # FIX: Added explicit filename and content-type 'video/mp4'
                files = {
                    'file': (filename, f, 'video/mp4')
                }
                headers = {"xi-api-key": self.api_key}

                data = {
                    'target_lang': target_language,
                    'num_speakers': 0  # Auto-detect
                }

                response = requests.post(
                    f"{self.base_url}/dubbing",
                    files=files,
                    data=data,
                    headers=headers,
                    timeout=self.timeout
                )

            if response.status_code == 200:
                video_id = response.json().get('dubbing_id')
                print(f"‚úÖ Video uploaded successfully!")
                print(f"Video ID: {video_id}")
                return video_id
            else:
                print(f"Error uploading video: {response.status_code}")
                print(f"Response: {response.text}")
                return None

        except Exception as e:
            print(f"Error uploading video: {e}")
            return None

    def start_dubbing(self, video_id: str, target_language: str, num_speakers: Optional[int] = None, watermark: bool = True) -> bool:
        try:
            headers = {"xi-api-key": self.api_key}
            payload = {
                "target_lang": target_language,
                "num_speakers": num_speakers if num_speakers else 0,
                "watermark": watermark
            }
            response = requests.post(
                f"{self.base_url}/dubbing/{video_id}/start",
                json=payload,
                headers=headers,
                timeout=self.timeout
            )
            if response.status_code in [200, 201]:
                print(f"üé¨ Dubbing started for language: {target_language}")
                return True
            else:
                print(f"Error starting dubbing: {response.status_code}")
                return False
        except Exception as e:
            print(f"Error starting dubbing: {e}")
            return False

    def check_dubbing_status(self, video_id: str) -> Optional[dict]:
        try:
            headers = {"xi-api-key": self.api_key}
            response = requests.get(
                f"{self.base_url}/dubbing/{video_id}",
                headers=headers,
                timeout=self.timeout
            )
            return response.json() if response.status_code == 200 else None
        except Exception as e:
            print(f"Error checking status: {e}")
            return None

    def wait_for_dubbing_completion(self, video_id: str, max_wait_minutes: int = 120) -> bool:
        start_time = time.time()
        max_wait_seconds = max_wait_minutes * 60
        poll_interval = 15
        print(f"‚è≥ Waiting for dubbing to complete...")

        while time.time() - start_time < max_wait_seconds:
            status = self.check_dubbing_status(video_id)
            if status is None: return False

            dub_status = status.get('status')
            elapsed = int(time.time() - start_time)
            print(f"[{elapsed//60}m {elapsed%60}s] Status: {dub_status}", end='\r')

            if dub_status == 'dubbed':
                print(f"\n‚úÖ Dubbing completed!")
                return True
            elif dub_status == 'failed':
                print(f"\n‚ùå Dubbing failed!")
                return False
            time.sleep(poll_interval)
        return False

    def get_dubbed_video(self, video_id: str, language_code: str, output_path: str) -> bool:
        try:
            headers = {"xi-api-key": self.api_key}
            response = requests.get(
                f"{self.base_url}/dubbing/{video_id}/audio/{language_code}",
                headers=headers,
                timeout=self.timeout,
                stream=True
            )
            if response.status_code == 200:
                os.makedirs(os.path.dirname(output_path) or '.', exist_ok=True)
                with open(output_path, 'wb') as f:
                    for chunk in response.iter_content(chunk_size=8192):
                        if chunk: f.write(chunk)
                print(f"‚úÖ Downloaded {language_code}: {output_path}")
                return True
            return False
        except Exception as e:
            print(f"Error downloading: {e}")
            return False

    def dub_video_complete(self, video_path: str, target_languages: List[str], output_dir: str = "/content/dubbed_videos", num_speakers: Optional[int] = None) -> dict:
        results = {'success': False, 'video_id': None, 'dubbed_videos': {}, 'errors': []}
        print("\n" + "="*60 + "\nüé¨ ELEVEN LABS VIDEO DUBBING\n" + "="*60)

        if not self.check_api_key():
            results['errors'].append("Invalid API key")
            return results

        video_id = self.upload_video(video_path, target_languages[0])
        if not video_id:
            results['errors'].append("Failed to upload video")
            return results

        results['video_id'] = video_id
        os.makedirs(output_dir, exist_ok=True)

        for idx, language in enumerate(target_languages, 1):
            print(f"\n3Ô∏è‚É£.{idx} Processing: {language}")
            if idx > 1:
                if not self.start_dubbing(video_id, language, num_speakers):
                    results['errors'].append(f"Failed to start {language}")
                    continue

            if self.wait_for_dubbing_completion(video_id):
                output_path = os.path.join(output_dir, f"dubbed_{language}.mp4")
                if self.get_dubbed_video(video_id, language, output_path):
                    results['dubbed_videos'][language] = output_path
            else:
                results['errors'].append(f"Dubbing failed for {language}")

        results['success'] = len(results['dubbed_videos']) > 0
        return results

print("‚úÖ ElevenLabsDubber class defined successfully!")
```
